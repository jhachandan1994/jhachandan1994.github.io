# Image classification using inception v3 model

```python
#Importing the required libraries
from tensorflow.keras.applications.inception_v3 import InceptionV3
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
from tensorflow.keras.applications.inception_v3 import preprocess_input, decode_predictions
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Downloading the model weights
model = InceptionV3(weights='imagenet')
```

```python
# save fine-tuning model for offline prediction purposes
model.save('inception_v3_imagenet_model.h5')
```

```python
def pred(img_for_recog):
    img = image.load_img(img_for_recog, target_size = (299, 299, 3))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    preds = model.predict(x)
    # decode the results into a list of tuples (class, description, probability)
    # (one such list for each sample in the batch)
    print('Predicted:', decode_predictions(preds, top = 5)[0])
```

```python
def display_img(img_for_display):
    img = load_img(img_for_display)
    plt.figure(figsize = (3,3))
    plt.imshow(img)
    plt.axis('off')
    plt.show()
```

```python
img_path = (r"C:\Users\jhach\Desktop\classification images\elephant.jpg")
display_img(img_path)
pred(img_path)
```

Model classifies the above given image as "Indian_elephant" with an accuracy of 94 %
